```python
# import tensorflow as tf
import pandas as pd
import tensorflow.compat.v1 as tf # 1.XX version 을 사용 
tf.disable_v2_behavior() # 2.xx version 기능을 사용하지 않겠음
```

```python
x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]
y_data = [152., 185., 180., 196., 142.]

x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
w1 = tf.Variable(tf.random_normal([1]))
w2 = tf.Variable(tf.random_normal([1]))
w3 = tf.Variable(tf.random_normal([1]))
b = tf.Variable(tf.random_normal([1]))

hf = x1*w1+x2*w2+x3*w3+b
# 예측값과 실제값 사이의차에대한 제곱합의 평균
cost = tf.reduce_mean(tf.square(hf -y))

opt = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = opt.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(3001):
    _, cv, hfv = sess.run([train, cost, hf], feed_dict = {x1: x1_data, x2:x2_data, x3:x3_data, y:y_data})
    if step%10 == 0:
        print(step, 'cost :', cv, 'prediction :', hfv)

```

```python
# x1, x2, x3가 모두 90일 때 모의고사의 점수 예측
print('90, 90, 90 : ', sess.run(hf, feed_dict = {x1: 90, x2:90, x3:90}))

# x1, x2, x3가 모두 50일 때 모의고사의 점수 예측
print('50, 50, 50 : ', sess.run(hf, feed_dict = {x1: 50, x2:50, x3:50}))
```

```python
x_data = [[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]]
y_data = [[152.],
          [185.],
          [180.],
          [196.],
          [142.]]

# 행렬의 곱셈의 형식으로 나타내면 수식이 간단해짐
```

```python
x = tf.placeholder(tf.float32, [None, 3])
y = tf.placeholder(tf.float32, [None, 1])
```

# Metrix Multiple
행렬의 곱셈의 형식으로 나타내면 수식이 간단해짐

```python
w = tf.Variable(tf.random_normal([3,1]))
b = tf.Variable(tf.random_normal([1]))
```

```python
hf = tf.matmul(x, w)+b #hf = x*w +b
# (none, 3) * (3, 1) -> y랑 같은 shape로 만들어줘야함
```

```python
cost = tf.reduce_mean(tf.square(hf-y))
opt = tf.train.GradientDescentOptimizer(1e-5)
train = opt.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(3001):
    _, cv, hfv = sess.run([train, cost, hf], feed_dict = {x: x_data, y:y_data})
    if step%10 == 0:
        print(step, 'cost :', cv, 'prediction :', hfv)
```

```python
# 모두 90일때 출력해보시오.
print(sess.run(hf, feed_dict={x : [[90, 90, 90]]}))
```

# 시합 합불 분류 모델만들기 

```python
x_data = [[1, 2],
          [2, 3],
          [3, 1],
          [4, 3],
          [5, 3],
          [6, 2]]
y_data = [[0],
          [0],
          [0],
          [1],
          [1],
          [1]]
```

# 로지스틱 회귀

```python
import numpy as np
np.shape(x_data)
np.shape(y_data)
```

```python
x = tf.placeholder(tf.float32, shape=[None, 2])
y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
w = tf.Variable(tf.random_normal([2,1]))
b = tf.Variable(tf.random_normal([1]))
```

```python
hf = tf.sigmoid(tf.matmul(x,w) + b) # 0~1사이로 나옴 
```

```python
cost = -tf.reduce_mean(y*tf.log(hf)+(1-y)*tf.log(1-hf))
```

```python
train = tf.train.GradientDescentOptimizer(0.01).minimize(cost)
```

```python
predicted = tf.cast(hf > 0.5, dtype= tf.float32) # True -> 1.0, False -> 0.0
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, y), dtype=tf.float32))
```

#  with 구문을 사용해서 세션만들기 

```python
with tf.Session() as sess: # sess = tf.Session() 과 같은 구문 -> sess.close. 들여쓰기가 되어있는 부분이 유효한 부분
    sess.run(tf.global_variables_initializer())
    for step in range(10001):
        _, cv = sess.run([train, cost], feed_dict = {x:x_data, y:y_data})
        if step%200 == 0:
            print(step, cv) 
    h, p, a = sess.run([hf, predicted, accuracy], feed_dict = {x:x_data, y:y_data})
    print('hf : ', h, 'pred : ', p , 'acc : ', a  )
```
