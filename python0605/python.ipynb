```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

fig = plt.figure()
x = np.arange(1,5)
a =fig.subplots(2,2)
a[0][0].plot(x, x*x)
a[0][0].grid(True)
a[0][0].set_title("square")
a[0][1].plot(x, np.sqrt(x))
a[0][1].set_title("square root")
a[0][1].grid(color = 'blue', ls = '-.', lw =1) #ls  , lw =   
a[1][0].plot(x, np.exp(x))
a[1][0].set_title("exp")
a[1][1].plot(x, np.log10(x))
a[1][1].set_title("log")

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

x5 = np.arange(1,10)
y5 = np.sin(x5)

plt.plot(x5, y5, 'b')
plt.title("sine")
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

x5 = np.arange(0, math.pi*2, 0.05)
y5 = np.sin(x5)

plt.xlim(0, 10)
plt.ylim(0, 1,5)
plt.xticks([0,2,4,6,8])
plt.yticks([-1,0,1])
plt.plot(x5, y5)
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
a5 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
a5.plot(x5, np.exp(x5))
a6 =a5.twinx()
a6.plot(x5, np.log(x5), 'rx')
a5.set_ylabel('log')
fig.legend(labels=('exp', 'log'), loc = 'upper left')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = [[30, 25, 50, 20], [40, 23, 51, 17], [ 35,22, 45, 19]]
x6 = np.arange(4)
fig = plt.figure()
ax6 = fig.add_axes([0,0,1,1])
ax6.bar(x6 + 0.00 ,data[0], color = 'lightpink', width = 0.25)
ax6.bar(x6 + 0.25 , data[1], color = 'm', width = 0.25)
ax6.bar(x6 + 0.50 , data[2], color = 'r', width = 0.25)
fig.legend(labels = ('develop', 'dba', 'system engineer'), loc = 'best')
```

```python
n = 5 
menMean = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)

ind = np.arange(n)

plt.bar(ind, menMean, width=0.35, color = 'lightblue')
plt.bar(ind, womenMeans, width=0.35,bottom = menMean, color = 'lightpink')

plt.xticks(ind, ('G1', 'G2','G3','G4','G5' ))
plt.ylabel('Score')
plt.yticks(np.arange(0, 81, 10))
plt.legend(labels =['Men', 'Women'])
plt.show()
```
