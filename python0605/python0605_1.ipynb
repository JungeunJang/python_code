```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

x = np.arange(0, math.pi * 2, 0.05)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("sine graph")
plt.show()
```

```python
plt.title("sin, cos")
plot(x, sin(x), "r")
plot(x, cos(x), "y")
plot(x, -sin(x), "g")
plot(x, -cos(x), "m")
```

```python
x1 = np.arange(0, math.pi*2, 0.05)
y1 = np.sin(x1)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) #left, bottom, width, 
ax.plot(x1,y1, "m")
ax.legend(labels=('x'), loc = 'lower right')
ax.set_title("Sine graph")
ax.set_xlabel("X axis")
ax.set_ylabel("y axis")
plt.show()
```
