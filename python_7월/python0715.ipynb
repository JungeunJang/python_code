```python
# import tensorflow as tf
import pandas as pd
import numpy as np
import tensorflow.compat.v1 as tf # 1.XX version 을 사용 
tf.disable_v2_behavior() # 2.xx version 기능을 사용하지 않겠음
```

```python
data = pd.read_csv('C:/Python37_Project/딥러닝/test-score.csv', names= ['x1', 'x2', 'x3', 'y'])
```

```python
x_data = np.array(data[['x1', 'x2', 'x3']])
y_data = np.array(data[['y']])
```

```python
x = tf.placeholder(tf.float32, [None, 3])
y = tf.placeholder(tf.float32, [None, 1])
```

```python
y_data
```

```python
w = tf.Variable(tf.random_normal([3,1]))
b = tf.Variable(tf.random_normal([1]))
```

```python
hf = tf.matmul(x, w)+b
cost = tf.reduce_mean(tf.square(hf-y))
opt = tf.train.GradientDescentOptimizer(1e-5)
train = opt.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(3001):
    _, cv, hfv = sess.run([train, cost, hf], feed_dict = {x: x_data, y: y_data})
    if step%10 == 0:
        print(step, 'cost :', cv, 'prediction :', hfv)
```
