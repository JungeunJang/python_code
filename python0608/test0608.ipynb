```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys

# 한글문제 해결처리 
################################################################
from matplotlib import font_manager, rc
import platform
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
################################################################

data = pd.read_csv("../data/21.모범음식점현황(사전정보공표181231).csv", encoding='cp949')
data.describe()
data2 = data.groupby(by='업태').size().reset_index(name='건수')
data3 = data2.sort_values(by='건수')
plt.plot(data3['업태'], data3['건수'])
plt.title("모범음식점 현황")
plt.xlabel('업태')
plt.xlabel('건수')
plt.show()
```

```python
# “거창군”과 “함양군”에서 하위 4곳의 음식점을 찾고 4곳의 음식점 기준으로 그래프를 출력하시오.(주메뉴기준)
filter1 = data['시군구'] == '거창군'
filter2 = data['시군구'] == '함양'

data4 =data[filter1]
data5 =data[filter2]

data_result = pd.concat([data4, data5])
data_result2 = data_result.groupby(by='주메뉴').size().reset_index(name='건수')
data_result3 = data_result2.sort_values(by=['건수','주메뉴'],  ascending = [False, True]).tail(4)
data_result3
plt.plot(data_result3['주메뉴'], data_result3['건수'])
plt.show()
```
