```python
import nltk
from wordcloud import wordcloud, STOPWORDS
from konlpy.corpus import kobill
file_ko = kobill.fileids()
doc_ko = kobill.open('1809890.txt').read()
doc_ko

from konlpy.tag import Twitter; t = Twitter()
tokens_ko = t.nouns(doc_ko)
tokens_ko
ko = nltk.Text(tokens_ko, name = '대한민국 국회의안 제 1809890호')
print(len(ko.tokens))
print(len(set(ko.tokens)))
ko.vocab()

import platform

path = "c:/Wimdows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc
import platform
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')

plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()

stop_word = ['.', ')', '(', ',', "'", '%', '-', 'X', ').', 'x','의', '자', '에', '안', '번', '호', '을', '이', '다', '만', '로', '가', '를', '제', '것임', '곧']
ko = [each_word for each_word in ko if each_word not in stop_word]
ko
ko = nltk.Text(ko, name = '대한민국 국회 의안 제 1909890호')

plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()

ko.count('초등학교')
plt.figure(figsize=(12,6))
ko.dispersion_plot(['육아휴작','초등학교','공무원'])
```

```python
from wordcloud import WordCloud

ko.concordance('초등학교')
ko.collocations()

data = ko.vocab().most_common(150)
font_path = 'c:/Wimdows/Fonts/malgun.ttf'
wordcloud = WordCloud(relative_scaling=0.2, background_color='white').generate_from_frequencies(dict(data))
plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
